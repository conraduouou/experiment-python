```python
from selenium import webdriver
import time

DRIVER_PATH = "C:\Development\msedgedriver.exe"

driver = webdriver.Edge(executable_path=DRIVER_PATH)

driver.get("https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors")
time.sleep(2)

driver.maximize_window()

with open("updated_salaries.csv", "w") as file:
    table = driver.find_element_by_class_name('data-table')

    to_write = ",".join( \
        element.text for element in table.find_elements_by_class_name('data-table__header')
    )

    file.write(to_write + "\n")

    for row in table.find_elements_by_class_name('data-table__row'):
        cells = []
        for cell in row.find_elements_by_class_name('data-table__cell'):
            if cell.text.startswith('$'):
                cells.append("".join(cell.text.replace("$", "").split(',')))
            else:
                cells.append(cell.text)
            
        print(cells)

        file.write(",".join(cells) + "\n")
```

```python
import pandas as pd

df = pd.read_csv('updated_salaries.csv')

df.insert(4, 'Mean Career Pay', df['Mid-Career Pay'].subtract(df['Early Career Pay']))

df
```
